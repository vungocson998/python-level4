```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

img1 = np.array([[0, 255],
                 [0, 255]])

plt.imshow(img1, cmap='gray')
plt.show()
```

```python
img1_size = 2*2
img1_fft = np.fft.fft2(img1)/img1_size
img1_fft = np.abs(img1_fft)
print(img1_fft)
plt.imshow(img1_fft, cmap='gray')
plt.show()
```

```python
import cv2
img2 = cv2.imread('./img1.png', cv2.IMREAD_GRAYSCALE)
print(type(img2))
print(np.amax(img2))
plt.imshow(img2, cmap='gray')
plt.show()
```

```python
img2_size = 300*300
img2_fft = np.fft.fft2(img2)/img2_size
img2_fft = np.fft.fftshift(img2_fft)
img2_fft = np.abs(img2_fft)
magnitude_spectrum2 = np.log(img2_fft)
plt.imshow(magnitude_spectrum2, cmap='gray')
plt.show()
```
